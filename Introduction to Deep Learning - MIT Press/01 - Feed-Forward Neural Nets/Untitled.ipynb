### <span style="color: rgb(0, 105, 175);font-family: 'NewComputerModernMath';">Q2</span>

#### <span style="font-family: 'NewComputerModernMath';">_We simplify our Mnist computation by assuming our “image” has two binary-valued pixels, 0 and 1, there are no bias parameters, and we are performing a binary classification problem._

#### <span style="font-family: 'NewComputerModernMath';">_(a) Compute the forward-pass logits and probabilities when the pixel values are [0,1], and the weights are:_

$$
+.2 \quad -.3
$$

$$
-.1 \quad +.4
$$

#### <span style="font-family: 'NewComputerModernMath';">_Here w[i,j] is the weight on the connection between the ith pixel and the jth unit. E.g., w[0, 1] here is− .3._ 

#### <span style="font-family: 'NewComputerModernMath';">_(b) Assume the correct answer is 1 (not 0) and use a learning rate of 0.1. What is the loss? Also, compute ∆w0,0 on the backward pass._

##### **Step 1: Compute Logits**
Given pixel values **$ \mathbf{x} = [0,1] $** and weight matrix:

$$
\mathbf{W} =
\begin{bmatrix}
0.2 & -0.3 \\
-0.1 & 0.4
\end{bmatrix}
$$

The logits are computed as:

$$
l_j = b_j + \mathbf{x} \cdot \mathbf{w}_j
$$

Since biases are not specified, we assume **$ b_j = 0 $**. Thus:

$$
l = \mathbf{x} \cdot \mathbf{W} = [0,1] \cdot
\begin{bmatrix}
0.2 & -0.3 \\
-0.1 & 0.4
\end{bmatrix}
$$

$$
l_0 = (0 \cdot 0.2) + (1 \cdot -0.1) = -0.1
$$

$$
l_1 = (0 \cdot -0.3) + (1 \cdot 0.4) = 0.4
$$

So, the logits are:

$$
l = [-0.1, 0.4]
$$

##### **Step 2: Compute Probabilities**
Using the softmax function:

$$
p(a) = \sigma_a(l) = \frac{e^{l_a}}{\sum_i e^{l_i}}
$$

$$
e^{-0.1} = 0.9048, \quad e^{0.4} = 1.4918
$$

$$
\sum e^l = 0.9048 + 1.4918 = 2.3966
$$

$$
p(0) = \frac{0.9048}{2.3966} \approx 0.3775, \quad p(1) = \frac{1.4918}{2.3966} \approx 0.6225
$$

##### **Step 3: Compute Loss**
Using cross-entropy loss:

$$
X(\Phi, x) = -\ln p(a)
$$

Since the true label is $ a = 1 $:

$$
X(\Phi, x) = -\ln (0.6225) \approx 0.474
$$

##### **Step 4: Compute Weight Update**
Gradient of cross-entropy loss with respect to logits:

$$
\frac{\partial X(\Phi)}{\partial l_j} =
\begin{cases}
-(1 - p_j) & a = j \\
p_j & a \neq j
\end{cases}
$$

For $ l_0 $:

$$
\frac{\partial X(\Phi)}{\partial l_0} = 0.3775
$$

For $ l_1 $:

$$
\frac{\partial X(\Phi)}{\partial l_1} = 0.6225 - 1 = -0.3775
$$

Gradient of weights:

$$
\Delta w_{i,j} = -\mathcal{L} x_i \frac{\partial X(\Phi)}{\partial l_j}
$$

##### **Update for $ w_{1,0} $**
$$
\Delta w_{1,0} = -0.1 \times 0 \times 0.3775 = 0 \quad (\text{No update since } x_0 = 0)
$$

##### **Update for $ w_{0,0} $**
$$
\Delta w_{0,0} = -0.1 \times 1 \times 0.3775 = -0.03775
$$

$$
w_{0,0} = -0.1 - 0.03775 = -0.13775
$$

##### **Update for $ w_{1,1} $**
$$
\Delta w_{1,1} = -0.1 \times 1 \times (-0.3775) = 0.03775
$$

$$
w_{1,1} = 0.4 + 0.03775 = 0.43775
$$

##### **Update for $ w_{0,1} $**
$$
\Delta w_{0,1} = -0.1 \times 0 \times (-0.3775) = 0 \quad (\text{No update since } x_0 = 0)
$$

##### **Final Answer**
- **Logits:** $ l = [-0.1, 0.4] $
- **Probabilities:** $ p(0) = 0.3775 $, $ p(1) = 0.6225 $
- **Loss:** $ X(\Phi, x) \approx 0.474 $
- **Weight Updates:**
  - $ w_{0,0} = -0.13775 $ (updated)
  - $ w_{1,0} $ remains unchanged
  - $ w_{0,1} $ remains unchanged
  - $ w_{1,1} = 0.43775 $ (updated)

---

##### **What does $ w_{i,j} $ represent?**
The notation $ w_{i,j} $ refers to the weights in a neural network's weight matrix $ \mathbf{W} $, where:

- $ i $ indexes the input feature dimension (or neuron in the previous layer).
- $ j $ indexes the output class (or neuron in the current layer).

In this example:
- $ w_{0,0} $ is the weight connecting input feature $ x_0 $ to logit $ l_0 $.
- $ w_{1,0} $ is the weight connecting input feature $ x_1 $ to logit $ l_0 $.
- $ w_{0,1} $ is the weight connecting input feature $ x_0 $ to logit $ l_1 $.
- $ w_{1,1} $ is the weight connecting input feature $ x_1 $ to logit $ l_1 $.

This means that each column in $ \mathbf{W} $ corresponds to a different output class, while each row corresponds to an input feature.

$$
p(a) = \sigma_a(l) = \frac{e^{l_a}}{\sum_i e^{l_i}}

$$

$$

$$

---

##### **Step 3: Compute Loss**

$$
X(\Phi, x) = -\ln p(a)
$$

If the true label is $ a = 1 $:

$$
X(\Phi, x) = -\ln (0.5) \approx 0.693
$$

---

##### **Step 4: Compute Weight Update**
Gradient of cross-entropy loss with respect to logits:

$$
\frac{\partial X(\Phi)}{\partial l_j} =
\begin{cases}
-(1 - p_j) & a = j \\
p_j & a \neq j
\end{cases}
$$

For $ l_0 $:

$$
\frac{\partial X(\Phi)}{\partial l_0} = 0.5
$$

For $ l_1 $:

$$
\frac{\partial X(\Phi)}{\partial l_1} = 0.5 - 1 = -0.5
$$

Gradient of weights:

$$
\Delta w_{i,j} = -\mathcal{L} x_i \frac{\partial X(\Phi)}{\partial l_j}
$$

$$
\Delta w_{i,j} = 0.1 \times 0 \times (\text{some gradient}) = 0
$$

Thus, **no weight updates occur** because all input values are 0.

---

##### **Final Answer**

---

##### **Why No Updates?**
Since the input is **all zeros**, the gradients for the weights will always be **zero**, meaning that the model does not change its weights in this case.

### <span style="color: rgb(0, 105, 175);font-family: 'NewComputerModernMath';">Q4</span>

#### <span style="font-family: 'NewComputerModernMath';">_A fellow student asks you, “In elementary calculus we foundminima of a function by diﬀerentiating it, setting the resulting expression to zero, and then solving the equation. Since our loss function is diﬀerentiable, why don’t we do that rather than bothering with gradient descent?” Explain why this is not, in fact, possible._

1. **High-Dimensionality**: In machine learning, the loss function often depends on thousands or even millions of parameters. Solving $\nabla L(\theta) = 0$ directly would require solving a massive system of nonlinear equations, which is computationally infeasible.

2. **Non-Closed Form Solutions**: Many loss functions in machine learning, especially those involving deep neural networks, lead to complex, non-analytic equations. Unlike elementary calculus problems that often yield closed-form solutions, these equations typically cannot be solved explicitly.

3. **Non-Convexity**: Loss functions in machine learning often have many local minima, saddle points, and flat regions. Traditional methods for solving $ \nabla L(\theta) = 0 $ rely on assumptions like convexity, which rarely hold in deep learning models.

4. **Computational Constraints**: Even if solving $ \nabla L(\theta) = 0 $ were possible in theory, it would require symbolic or numerical solvers that scale poorly with large datasets and high-dimensional parameter spaces. Gradient descent, on the other hand, provides an iterative and scalable method that efficiently finds good minima.

5. **Approximation in Practice**: Machine learning is not about finding the *exact* minimum but rather a *good enough* solution that generalizes well to unseen data. Gradient descent and its variants (SGD, Adam, etc.) allow for flexible and efficient optimization even in cases where exact solutions are not feasible.

### <span style="color: rgb(0, 105, 175);font-family: 'NewComputerModernMath';">Q6</span>

#### <span style="font-family: NewComputerModernMath;">_In this chapter, we limited ourselves to classification problems, for which cross-entropy is typically the loss function of choice.   There are also problems where we want our NN to predict particular values.   For example, undoubtedly, many folks would like a program that, given the price of a particular stock today plus all sorts of other facts about the world, outputs the price of the stock tomorrow.   If we were training a single-layer NN to do this, we would typically use the squared-error loss:_

$$
L(\mathbf{X}, \Phi) = (t - l(\mathbf{X}, \Phi))^2 \quad \quad (1.32)
$$  



#### <span style="font-family: NewComputerModernMath;">_where $ t $ is the actual price achieved on that day, and $ l(\mathbf{X}, \Phi) $ is the output of the one-layer NN with $ \Phi = \{\mathbf{b}, \mathbf{W}\} $.  (This is also known as <i>quadratic loss</i>.)  Derive the equation for the derivative of the loss concerning $ b_i $._
</span>


##### Given:
The squared error loss function:
$$
L(\mathbf{X}, \Phi) = (t - l(\mathbf{X}, \Phi))^2
$$
where:
- $ t $ is the actual price.
- $ l(\mathbf{X}, \Phi) $ is the neural network output.
- The neural network is parameterized by $ \Phi = \{\mathbf{b}, \mathbf{W}\} $, where $ \mathbf{b} $ is the bias vector and $ \mathbf{W} $ is the weight matrix.

We need to compute the derivative of the loss with respect to $ b_i $, the $ i $-th component of the bias vector.

---

##### Step 1: Compute the derivative of the loss function
Using the chain rule:

$$
\frac{\partial L}{\partial b_i} = 2 (t - l(\mathbf{X}, \Phi)) \cdot \frac{\partial (-l(\mathbf{X}, \Phi))}{\partial b_i}
$$

which simplifies to:

$$
\frac{\partial L}{\partial b_i} = -2 (t - l(\mathbf{X}, \Phi)) \frac{\partial l(\mathbf{X}, \Phi)}{\partial b_i}
$$

---

##### Step 2: Compute $\frac{\partial l(\mathbf{X}, \Phi)}{\partial b_i} $
For a single-layer neural network, the output is typically:

$$
l(\mathbf{X}, \Phi) = \mathbf{W} \mathbf{X} + \mathbf{b}
$$

Taking the partial derivative with respect to $ b_i $:

$$
\frac{\partial l(\mathbf{X}, \Phi)}{\partial b_i} = 1
$$

---

##### Step 3: Final Expression
Substituting this into our previous equation:

$$
\frac{\partial L}{\partial b_i} = -2 (t - l(\mathbf{X}, \Phi))
$$

which simplifies to:

$$
\frac{\partial L}{\partial b_i} = 2 (l(\mathbf{X}, \Phi) - t)
$$

---

##### Final Answer:
$$
\frac{\partial L}{\partial b_i} = 2 (l(\mathbf{X}, \Phi) - t)
$$

This result shows that the gradient of the loss with respect to $ b_i $ is proportional to the difference between the predicted value and the actual value. This makes sense intuitively: if the prediction is larger than the actual value, the gradient is positive, meaning we should decrease $ b_i $ to reduce the loss, and vice versa.
