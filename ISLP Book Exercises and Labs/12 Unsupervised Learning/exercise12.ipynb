To prove the given equation:
$$
\frac{1}{|C_k|} \sum_{i,i' \in C_k} \sum_{j=1}^{p} (x_{ij} - x_{i'j})^2 = 2 \sum_{i \in C_k} \sum_{j=1}^{p} (x_{ij} - \bar{x}_{kj})^2
$$
where $ C_k $ is a cluster (or set of indices), $ |C_k| $ is the number of elements in $ C_k $, and $ \bar{x}_{kj} $ is the mean of the $ j $-th feature over the cluster, we proceed as follows.

\section*{Step 1: Expanding the Left-Hand Side (LHS)}

The left-hand side involves a double sum over all pairs $ i, i' \in C_k $, measuring pairwise squared differences:
$$
\sum_{i,i' \in C_k} \sum_{j=1}^{p} (x_{ij} - x_{i'j})^2.
$$

Expanding the squared term:
$$
(x_{ij} - x_{i'j})^2 = x_{ij}^2 - 2x_{ij} x_{i'j} + x_{i'j}^2.
$$

Summing over all pairs $ i, i' $:
$$
\sum_{i,i' \in C_k} \sum_{j=1}^{p} (x_{ij}^2 - 2x_{ij}x_{i'j} + x_{i'j}^2).
$$

Rewriting the summation:
$$
\sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{ij}^2 - 2\sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{ij}x_{i'j} + \sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{i'j}^2.
$$

Since the last sum is symmetric with the first, we rewrite:
$$
2 \sum_{j=1}^{p} \sum_{i \in C_k} |C_k| x_{ij}^2 - 2 \sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{ij} x_{i'j}.
$$

Dividing by $ |C_k| $, we obtain:
$$
\frac{2}{|C_k|} \sum_{j=1}^{p} \left( \sum_{i \in C_k} |C_k| x_{ij}^2 - \sum_{i,i' \in C_k} x_{ij} x_{i'j} \right).
$$

_Step 2: Expressing the Right-Hand Side (RHS)_

The right-hand side sums the squared deviations from the mean:
$$
2 \sum_{i \in C_k} \sum_{j=1}^{p} (x_{ij} - \bar{x}_{kj})^2.
$$

Expanding:
$$
2 \sum_{i \in C_k} \sum_{j=1}^{p} (x_{ij}^2 - 2x_{ij} \bar{x}_{kj} + \bar{x}_{kj}^2).
$$

Rewriting in summation form:
$$
2 \sum_{j=1}^{p} \sum_{i \in C_k} x_{ij}^2 - 4 \sum_{j=1}^{p} \sum_{i \in C_k} x_{ij} \bar{x}_{kj} + 2 \sum_{j=1}^{p} \sum_{i \in C_k} \bar{x}_{kj}^2.
$$

Since $ \bar{x}_{kj} = \frac{1}{|C_k|} \sum_{i \in C_k} x_{ij} $, we substitute:
$$
\sum_{i \in C_k} x_{ij} \bar{x}_{kj} = \sum_{i \in C_k} x_{ij} \left( \frac{1}{|C_k|} \sum_{i' \in C_k} x_{i'j} \right) = \frac{1}{|C_k|} \sum_{i,i' \in C_k} x_{ij} x_{i'j}.
$$

Thus, the RHS becomes:
$$
2 \sum_{j=1}^{p} \sum_{i \in C_k} x_{ij}^2 - \frac{4}{|C_k|} \sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{ij} x_{i'j} + 2 \sum_{j=1}^{p} |C_k| \bar{x}_{kj}^2.
$$

Since $ \sum_{i \in C_k} \bar{x}_{kj}^2 = |C_k| \bar{x}_{kj}^2 $, we simplify:
$$
2 \sum_{j=1}^{p} \sum_{i \in C_k} x_{ij}^2 - \frac{2}{|C_k|} \sum_{j=1}^{p} \sum_{i,i' \in C_k} x_{ij} x_{i'j}.
$$

Comparing both sides, we see that they are equal

#### <span style="font-family: NewComputerModernMath;">_(b) On the basis of this identity, argue that the **$ K $-means clustering algorithm** (Algorithm **12.2**) decreases the **objective function** (Equation **12.17**) at each iteration._


At each step of the **$K$-means clustering algorithm**, each observation is assigned to the nearest cluster center, reducing its distance to the centroid. Then, the centroids are updated to minimize the overall cluster variance. Since both steps either reduce or keep the objective function (Equation **12.17**) the same, the algorithm ensures that the objective function decreases at each iteration.

#### <span style="font-family: NewComputerModernMath;">In this problem, you will perform $ K $-means clustering manually, with $ K = 2 $, on a small example with $ n = 6 $ observations and $ p = 2 $ features. The observations are as follows.

$$
\begin{array}{|c|c|c|}
\hline
\text{Obs.} & X_1 & X_2 \\
\hline
1 & 1 & 4 \\
2 & 1 & 3 \\
3 & 0 & 4 \\

\hline

$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Define the dataset
data = np.array([
    [1, 4],
    [1, 3],
    [0, 4],
    [5, 1],
    [6, 2],
    [4, 0]
])

# Extracting X1 and X2 for plotting
X1 = data[:, 0]
X2 = data[:, 1]
```

#### <span style="font-family: NewComputerModernMath;">_(a) Plot the observations._  

```python
# Plot the observations
plt.figure(figsize=(6,6))
plt.scatter(X1, X2, c='black', marker='o', label="Observations")
plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.title("Scatter Plot of Observations")
plt.legend()
plt.grid()
plt.show()
```

#### <span style="font-family: NewComputerModernMath;">_(b) Randomly assign a cluster label to each observation. You can use the_  
    `np.random.choice()` function to do this. Report the cluster labels for each observation.  

```python
np.random.seed(42)  # Set seed for reproducibility
cluster_labels = np.random.choice([0, 1], size=len(data))
print("Initial Random Cluster Assignments:", cluster_labels)
```

#### <span style="font-family: NewComputerModernMath;">_(c) Compute the centroid for each cluster._  

```python
def compute_centroids(data, labels, k=2):
    centroids = np.zeros((k, data.shape[1]))
    for i in range(k):
        centroids[i] = data[labels == i].mean(axis=0)
    return centroids

cluster_labels = np.array(cluster_labels)

centroids = compute_centroids(data, cluster_labels)
print("Initial Centroids:\n", centroids)
```

#### <span style="font-family: NewComputerModernMath;">_(d) Assign each observation to the centroid to which it is closest, in terms of Euclidean distance. Report the cluster labels for each observation._  

```python
def assign_clusters(data, centroids):
    distances = np.linalg.norm(data[:, np.newaxis] - centroids, axis=2)
    return np.argmin(distances, axis=1)

new_labels = assign_clusters(data, centroids)
print("Updated Cluster Assignments:", new_labels)
```

#### <span style="font-family: NewComputerModernMath;">_(e) Repeat (c) and (d) until the answers obtained stop changing._  

```python
def k_means_clustering(data, k=2, max_iters=10):
    np.random.seed(42)
    labels = np.random.choice([0, 1], size=len(data))
    
    for _ in range(max_iters):
        centroids = compute_centroids(data, labels, k)
        new_labels = assign_clusters(data, centroids)
        
        if np.all(labels == new_labels):
            break  # Stop if clusters do not change
        
        labels = new_labels  # Update labels
    
    return labels, centroids

final_labels, final_centroids = k_means_clustering(data)

print("Final Cluster Assignments:", final_labels)
print("Final Centroids:\n", final_centroids)
```

#### <span style="font-family: NewComputerModernMath;">_(f) In your plot from (a), color the observations according to the cluster labels obtained._  
</span>

```python
# Plot final clusters
plt.figure(figsize=(6,6))

# Scatter plot with cluster colors
plt.scatter(X1, X2, c=final_labels, cmap='coolwarm', marker='o', edgecolors='k', label="Observations")

# Plot centroids
plt.scatter(final_centroids[:, 0], final_centroids[:, 1], c='coral', marker='X', s=200, label="Centroids")

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.title("Final Cluster Assignments")
plt.legend()
plt.grid()
plt.show()
```

In each regression, the principal component score vectors are the predictors, and one of the features of the data matrix is the response.

$$

$$

$$
\min_{\mathbf{A} \in \mathbb{R}^{n \times M}, \mathbf{B} \in \mathbb{R}^{p \times M}} 
\left\{ \sum_{j=1}^{p} \sum_{i=1}^{n} \left( x_{ij} - \sum_{m=1}^{M} a_{im} b_{jm} \right)^2 \right\}. \quad \quad \quad (12.6)
$$


If I understand the question correctly:

When we create principal component (PC) scores, we transform the original dataset into a smaller set of uncorrelated predictors derived from the original ones. If we want to approximate the original features using the PC scores and loading vectors, we can do so by performing 
𝑝 separate least squares regressions. In each regression, a feature from the original dataset is treated as the response, and the PC scores are used as predictors to estimate its values.

Each column $ j $ of **X** is a vector $ X_j $ (length $ n $), and it satisfies:

$$
X_j \approx \mathbf{Z} \phi_j.
$$

For each feature $ j $, we perform a **separate least squares regression**:

$$
\hat{\phi}_j = (\mathbf{Z}^T \mathbf{Z})^{-1} \mathbf{Z}^T X_j.
$$

Thus, the **principal component loading vectors** are obtained by solving $ p $ separate least squares problems, where:

- The predictors are the principal component scores $ \mathbf{Z} $.
- The response is one of the columns of the data matrix $ X_j $.

##### Why Standardization is Essential Before Clustering  

When performing clustering, it is crucial to **standardize** the data to ensure that all variables contribute equally to the analysis. This is because different predictors (parameters) may have different units of measurement, which can disproportionately influence the clustering results.  

##### **Example: Impact of Different Scales**  
Consider a dataset with the following variables:  
- **Column 1**: GPA (ranges from 1.0 to 4.0)  
- **Column 2**: Height (measured in centimeters, ranging from 150 to 230)  

Without standardization, when calculating distances between observations:  
- A large difference in **GPA** (e.g., 4.0 - 1.0 = 3.0) would have minimal impact on the clustering.  
- Meanwhile, a small difference in **height** (e.g., 190 cm - 184 cm = 6 cm) would dominate the distance computation due to its larger numerical range.  

Since clustering algorithms often rely on distance metrics like Euclidean distance, variables with larger magnitudes will have a greater influence, potentially leading to misleading clustering results.  

##### **The Solution: Standardization**  
By standardizing the data (e.g., using **z-score normalization**), we ensure that each variable has:  
- A mean of **0**  
- A standard deviation of **1**  

This transformation ensures that all variables contribute **equally** to the clustering process, leading to more meaningful and unbiased groupings.  

##### **Conclusion**  
To achieve accurate and reliable clustering results, **standardization is essential** when dealing with variables on different scales. This prevents certain features from disproportionately influencing the clustering structure, ensuring that all variables are treated fairly in the analysis

```python
import numpy as np
from sklearn.decomposition import PCA

def matrix_completion(X, rank=2, max_iter=100, tol=1e-4, alpha=0.5, verbose=False):
    X_filled = X.copy()
    missing_mask = np.isnan(X)
    
    # Step 1: Initialize missing values with column means
    col_means = np.nanmean(X, axis=0)
    X_filled[missing_mask] = np.take(col_means, np.where(missing_mask)[1])

    errors = []
    
    for iteration in range(1, max_iter + 1):
        # Step 2(a): Compute principal components of X_filled
        pca = PCA(n_components=rank)
        X_reconstructed = pca.fit_transform(X_filled) @ pca.components_

        # Step 2(b): Weighted update for missing values
        X_filled[missing_mask] = alpha * X_filled[missing_mask] + (1 - alpha) * X_reconstructed[missing_mask]

        # Step 2(c): Compute relative error
        observed_mask = ~missing_mask
        num = X[observed_mask] - X_filled[observed_mask]
        denom = np.linalg.norm(X[observed_mask])
        error = num
        errors.append(error)

        if verbose:
            print(f"Iteration {iteration}: Relative Error = {error:.6f}")

        # Check stopping condition
        if iteration > 1 and abs(errors[-1] - errors[-2]) < tol:
            if verbose:
                print("Convergence reached.")
            break

    return X_filled, errors
```

```python
def matrix_completion(X, rank=2, max_iter=100, tol=1e-4, verbose=False):
    X_filled = X.copy()
    missing_mask = np.isnan(X)
    
    col_means = np.nanmean(X, axis=0)
    X_filled[missing_mask] = np.take(col_means, np.where(missing_mask)[1])

    errors = []
    
    for iteration in range(1, max_iter + 1):
        # Perform SVD
        U, S, Vt = np.linalg.svd(X_filled, full_matrices=False)
        
        # Keep only the top-rank components
        S[rank:] = 0
        X_reconstructed = (U * S) @ Vt
        
        # Fill missing values
        X_filled[missing_mask] = X_reconstructed[missing_mask]  
        
        # Compute error on observed values
        observed_mask = ~missing_mask
        error = np.sum((X_filled[observed_mask] - X_reconstructed[observed_mask])**2)
        errors.append(error)

        if verbose:
            print(f"Iteration {iteration}: Relative Error = {error:.6f}")

        if iteration > 1 and abs(errors[-1] - errors[-2]) < tol:
            if verbose:
                print("Convergence reached.")
            break

    return X_filled, np.mean(errors)

# Experiment parameters
missing_fractions = np.arange(0.05, 0.35, 0.05)  # 5% to 30% in steps of 5%
print("missing fractions ratios: ",missing_fractions)
M_values = np.arange(1, 14)  # M = 1, ..., 8
num_repetitions = 10

# Store results
results = []

# Convert Boston DataFrame to NumPy
X_original = Boston

for frac in missing_fractions:
    for M in M_values:
        errors = []

        for _ in range(num_repetitions):
            # Create a copy of the data
            X = X_original.copy()

            # Randomly mask a fraction of the data
            mask = np.random.rand(*X.shape) < frac
            X[mask] = np.nan

            # Apply matrix completion
            _, error = matrix_completion(X, rank=M)
            errors.append(error)

        # Store the average error
        avg_error = np.mean(errors)
        results.append([frac, M, avg_error])

# Convert results to DataFrame
results_df = pd.DataFrame(results, columns=["Missing Fraction", "M", "Error"])

# Pivot for visualization
error_matrix = results_df.pivot(index="Missing Fraction", columns="M", values="Error")

# Plot results
plt.figure(figsize=(10, 6))
for M in M_values:
    plt.plot(error_matrix.index, error_matrix[M], marker='o', label=f'M={M}')

plt.xlabel("Fraction of Missing Data")
plt.ylabel("Approximation Error")
plt.title("Matrix Completion Error vs. Missing Data Fraction & Rank M")
plt.legend(title="Rank M")
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import scipy.cluster.hierarchy as sch
import scipy.spatial.distance as ssd
from scipy.stats import ttest_ind
from statsmodels.stats.multitest import multipletests
import matplotlib.pyplot as plt

corr_matrix = np.corrcoef(df.T)  # Compute sample correlation
corr_distance = 1 - corr_matrix  # Convert to distance metric
condensed_dist = ssd.squareform(corr_distance, checks=False)

linkage_matrix = sch.linkage(condensed_dist, method="complete")

from scipy.cluster.hierarchy import fcluster
cluster_labels = fcluster(linkage_matrix, 2, criterion='maxclust')

sample_groups = {sample: cluster for sample, cluster in zip(df.columns, cluster_labels)}

# Split data into two groups based on clustering
group1_samples = [s for s, g in sample_groups.items() if g == 1]
group2_samples = [s for s, g in sample_groups.items() if g == 2]

df_group1 = df[group1_samples]
df_group2 = df[group2_samples]

p_values = []
fold_changes = []

for gene in df.index:
    t_stat, p_val = ttest_ind(df_group1.loc[gene], df_group2.loc[gene], equal_var=False)
    p_values.append(p_val)
    
    
    mean_group1 = df_group1.loc[gene].mean()
    mean_group2 = df_group2.loc[gene].mean()
    fold_change = np.log2((mean_group1 + 1e-6) / (mean_group2 + 1e-6))  # Adding small value to avoid division by zero
    fold_changes.append(fold_change)

results_df = pd.DataFrame({
    'Gene': df.index,
    'p_value': p_values,
    'log2FC': fold_changes
})

# Apply multiple testing correction (FDR)
results_df['adjusted_p'] = multipletests(results_df['p_value'], method='fdr_bh')[1]

results_df = results_df.sort_values(by=['adjusted_p', 'log2FC'], ascending=[True, False])

display(results_df)
```
