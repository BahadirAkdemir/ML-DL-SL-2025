#### <span style="font-family: TheModernComputerMath;">*(a) In a study of disease relapse, due to a careless research scientist, all patients whose phone numbers begin with the number “2” are lost to follow up.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **independent** because the loss to follow-up is based on an arbitrary factor (phone number) unrelated to disease relapse.

#### <span style="font-family: TheModernComputerMath;">*(b) In a study of longevity, a formatting error causes all patient ages that exceed 99 years to be lost (i.e., we know that those patients are more than 99 years old, but we do not know their exact ages).*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because patients who live longer than 99 years are systematically censored, meaning the probability of being censored is directly related to the outcome of interest (longevity).

#### <span style="font-family: TheModernComputerMath;">*(c) Hospital A conducts a study of longevity. However, very sick patients tend to be transferred to Hospital B, and are lost to follow up.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because very sick patients, who likely have shorter survival times, are more likely to be lost to follow-up, making the probability of being censored related to the outcome of interest (longevity).

#### <span style="font-family: TheModernComputerMath;">*(d) In a study of unemployment duration, the people who find work earlier are less motivated to stay in touch with study investigators, and therefore are more likely to be lost to follow up.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because individuals who find work earlier (shorter unemployment duration) are more likely to be lost to follow-up, making the probability of being censored related to the outcome of interest (unemployment duration).

#### <span style="font-family: TheModernComputerMath;">*(e) In a study of pregnancy duration, women who deliver their babies pre-term are more likely to do so away from their usual hospital, and thus are more likely to be censored, relative to women who deliver full-term babies.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because women who deliver pre-term (shorter pregnancy duration) are more likely to be censored, making the probability of being censored related to the outcome of interest (pregnancy duration).

#### <span style="font-family: TheModernComputerMath;">*(f) A researcher wishes to model the number of years of education of the residents of a small town. Residents who enroll in college out of town are more likely to be lost to follow up, and are also more likely to attend graduate school, relative to those who attend college in town.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because individuals who leave town for college are more likely to be censored and also tend to have longer education durations, making the probability of being censored related to the outcome of interest (years of education).

#### <span style="font-family: TheModernComputerMath;">*(g) Researchers conduct a study of disease-free survival (i.e., time until disease relapse following treatment). Patients who have not relapsed within five years are considered to be cured, and thus their survival time is censored at five years.*</span>

&nbsp;&nbsp;&nbsp;&nbsp; The censoring mechanism is **dependent (informative censoring)** because patients who are censored at five years are those who have not relapsed, meaning the probability of being censored is directly related to the outcome of interest (disease-free survival).

$$
0.8 \times 0.75 = 0.6.

As the question says, if there is no difference between the survival functions of the two groups. Then we can think of $ q_{1k} $ as the number of failures if we draw $ r_{1k} $ observations, without replacement, from a risk set of $ r_k $ observations that contains a total of $ q_k $ failures. Now let's see how a hypergeometric distribution behaves:

_**The Setting**_
- Two Groups: We have Group 1 and Group 2.
- **At-Risk Individuals**: at the $k^{th}$ event time $d_k$, there are $r_k$ total individuals still at risk (combined across both groups).
  - $r_{1k}$ for Group 1 and $r_{2k}$ for Group 2 ($r_{k}$ = $r_{1k}$ + $r_{2k}$)
- **Events at $d_k$**: Exactly $q_k$ events (i.e., “failures”) occur at time $d_k$, among the $r_k$ at‐risk individuals.
- **Number of Events in Group 1**: Let $q_{1k}$ be the number of these $q_k$ events that happen to individuals in Group 1.

_**Connecting to the Hypergeometric Distribution**_

The hypergeometric distribution describes the following classical scenario:

- We have a population of size 𝑁 
- Among these 𝑁 items, 𝐾 are labeled “successes” (and 𝑁-𝐾 are “failures”).
- We draw 𝑛 items without replacement from this population of 𝑁
- The random variable 𝑋 is the number of successes in our draw of 𝑛

In symbols, 𝑋 ∼ $Hypergeometric(N,K,n)$

_**Mapping Our Problem to the Hypergeometric Setup**_

1. **Population Size** $ N = r_k $.  
   The "population" is the set of all at-risk individuals at time $ d_k $.

2. **Number of Marked (Group 1) Individuals** $ K = r_{k1} $.  
   Think of Group 1 individuals as "marked" or "successes."

3. **Draw Size** $ n = q_k $.  
   We are effectively "drawing" $ q_k $ individuals from the $ r_k $ who will experience the event at $ d_k $.

4. **Number of Marked Individuals in the Draw** $ X = q_{1k} $.  
   This is precisely the number of events that happen to Group 1 out of the total $ q_k $ events.

Hence, under the null hypothesis (i.e., events are assigned uniformly at random among the $ r_k $ at-risk subjects), $ q_{1k} $ has a hypergeometric distribution:

$$
q_{1k} \sim \text{Hypergeometric} \left( \underbrace{r_k}_{N}, \underbrace{r_{1k}}_{K}, \underbrace{q_k}_{n} \right).
$$



#### <span style="font-family: TheModernComputerMath;">_**(b)** Given your previous answer, and the properties of the hypergeometric distribution, what are the mean and variance of $ q_{1k} $? Compare your answer to (11.5) and (11.6)._</span>

_**Mean and Variance**_

Once we identify $ q_{1k} $ as 

$$
\text{Hypergeometric}(r_k, r_{k1}, q_k),
$$

the mean and variance follow directly from the known formulas for the hypergeometric distribution:

*1. Mean:*

The general formula for the expectation of a hypergeometric distribution is:

$$
\mathbb{E}[X] = K \frac{n}{N}
$$

For our specific case:

$$
\mathbb{E}[q_{1k}] = r_{k1} \frac{q_k}{r_k}.
$$

*2. Variance:*

The general variance formula for a hypergeometric distribution is:

$$
\text{Var}(X) = K \frac{n}{N} \left(1 - \frac{n}{N} \right) \frac{N - K}{N - 1}.
$$

For our specific case:

$$
\text{Var}(q_{1k}) = r_{k1} \frac{q_k}{r_k} \left(1 - \frac{q_k}{r_k} \right) \frac{r_k - r_{k1}}{r_k - 1}.
$$

These are exactly the expressions that appear in the log-rank test derivations (equations **(11.5) and (11.6)**), confirming that the log-rank test follows the hypergeometric distribution.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

km = KaplanMeierFitter()
km_brain = km.fit(BrainCancer['time'], BrainCancer['status'])

# Create the plot
fig, ax = plt.subplots(figsize=(8, 8))
km_brain.plot(label='Kaplan-Meier estimate', ax=ax)

# Extract survival function and event table
surv_func = km_brain.survival_function_
timepoints = surv_func.index
survival_prob = surv_func.values.flatten()

# Greenwood's formula for standard error
event_table = km_brain.event_table
n_risk = event_table.loc[timepoints, 'at_risk']
n_event = event_table.loc[timepoints, 'observed']

# Compute variance of the survival function using Greenwood's formula
variance = (survival_prob**2) * np.cumsum(n_event / (n_risk * (n_risk - n_event)))
std_err = np.sqrt(variance)  # Standard error

# Compute the ±1 SE curves
upper_curve = np.minimum(survival_prob + std_err, 1)  # Ensure max 1
lower_curve = np.maximum(survival_prob - std_err, 0)  # Ensure min 0

# Plot the ±1 SE bounds
ax.plot(timepoints, upper_curve, 'r--', label='+1 SE')
ax.plot(timepoints, lower_curve, 'r--', label='-1 SE')

# Show the legend and plot
plt.legend()
plt.show()
```

####  <span style="font-family: TheModernComputerMath;">*(b) Draw a bootstrap sample of size $ n = 88 $ from the pairs $ (y_i, \delta_i) $, and compute the resulting Kaplan-Meier survival curve. Repeat this process $ B = 200 $ times. Use the results to obtain an estimate of the standard error of the Kaplan-Meier survival curve at each timepoint. Compare this to the standard errors obtained in (a).*

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from lifelines import KaplanMeierFitter

# Load the dataset (assuming BrainCancer dataframe exists)
n = 88  # Bootstrap sample size
B = 200  # Number of bootstrap samples
km = KaplanMeierFitter()

# Fit KM to the original dataset to get reference timepoints
km.fit(BrainCancer['time'], BrainCancer['status'])
timepoints = km.survival_function_.index  # Reference time grid

# Store bootstrap survival estimates at each timepoint
bootstrap_survival_curves = np.zeros((B, len(timepoints)))

for i in range(B):
    # Bootstrap sample (resample with replacement)
    bootstrap_sample = BrainCancer.sample(n=n, replace=True)

    # Fit Kaplan-Meier estimator to the bootstrap sample
    km.fit(bootstrap_sample['time'], bootstrap_sample['status'])
    
    # Reindex to match reference timepoints (use forward-fill interpolation)
    survival_interp = km.survival_function_.reindex(timepoints, method='ffill')

    # Store interpolated survival estimates
    bootstrap_survival_curves[i, :] = survival_interp.values.flatten()

# Compute the mean survival estimate across bootstrap samples
mean_survival = np.mean(bootstrap_survival_curves, axis=0)

# Compute the standard error across bootstrap samples
bootstrap_std_err = np.std(bootstrap_survival_curves, axis=0)

# Extract original survival estimate from part (a) and interpolate it
km_brain = km.fit(BrainCancer['time'], BrainCancer['status'])
surv_func = km_brain.survival_function_
survival_prob = surv_func.reindex(timepoints, method='ffill').values.flatten()  # Ensure aligned with timepoints

# Plot Kaplan-Meier survival curve with bootstrap standard errors
fig, ax = plt.subplots(figsize=(8, 8))
ax.step(timepoints, survival_prob, where="post", label="Kaplan-Meier Estimate", linewidth=2)

# Compute ±1 SE bounds from bootstrap method
upper_bootstrap = np.minimum(mean_survival + bootstrap_std_err, 1)
lower_bootstrap = np.maximum(mean_survival - bootstrap_std_err, 0)

# Plot bootstrap confidence bands
ax.plot(timepoints, upper_bootstrap, 'g--', label='Bootstrap +1 SE')
ax.plot(timepoints, lower_bootstrap, 'g--', label='Bootstrap -1 SE')

# Compare with Greenwood’s standard errors from (a)
event_table = km_brain.event_table
n_risk = event_table.loc[timepoints, 'at_risk']
n_event = event_table.loc[timepoints, 'observed']
variance = (survival_prob ** 2) * np.cumsum(n_event / (n_risk * (n_risk - n_event)))
std_err = np.sqrt(variance)  # Greenwood's Standard error

upper_greenwood = np.minimum(survival_prob + std_err, 1)
lower_greenwood = np.maximum(survival_prob - std_err, 0)

ax.plot(timepoints, upper_greenwood, 'r--', label='part(a) +1 SE')
ax.plot(timepoints, lower_greenwood, 'r--', label='part(a) -1 SE')

# Labels and legend
ax.set_xlabel("Time (days)")
ax.set_ylabel("Survival Probability")
plt.title("Kaplan-Meier Survival Curve with Bootstrap & Greenwood SE")
plt.legend()
plt.show()
```
