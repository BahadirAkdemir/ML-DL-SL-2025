<span style="font-family: 'NewComputerModernMath';">
1. It was mentioned in this chapter that a cubic regression spline with one knot at $ \xi $ can be obtained using a basis of the form $ x, x^2, x^3, (x - \xi)_+^3 $, where $ (x - \xi)_+^3 = (x - \xi)^3 $ if $ x > \xi $ and equals 0 otherwise. We will now show that a function of the form:

$$
f(x) = \beta_0 + \beta_1 x + \beta_2 x^2 + \beta_3 x^3 + \beta_4 (x - \xi)_+^3
$$

is indeed a cubic regression spline, regardless of the values of $ \beta_0, \beta_1, \beta_2, \beta_3, \beta_4 $.

<span style="font-family: 'NewComputerModernMath';">(a) Find a cubic polynomial:

$$
f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3
$$

<span style="font-family: 'NewComputerModernMath';">such that $ f(x) = f_1(x) $ for all $ x \leq \xi $. Express $ a_1, b_1, c_1, d_1 $ in terms of $ \beta_0, \beta_1, \beta_2, \beta_3, \beta_4 $.

<span style="font-family: 'NewComputerModernMath';"> Since $ x \leq \xi $, the fifth term ($ \beta_4 (x - \xi)_+^3$) will not be added to equation. Therefore:

$$
a_1 = \beta_0,\space b_1 = \beta_1,\space c_1 = \beta_2,\space d_1 = \beta_3
$$

<span style="font-family: 'NewComputerModernMath';">(b) Find a cubic polynomial:

$$
f_2(x) = a_2 + b_2 x + c_2 x^2 + d_2 x^3
$$

<span style="font-family: 'NewComputerModernMath';">such that $ f(x) = f_2(x) $ for all $ x > \xi $. Express $ a_2, b_2, c_2, d_2 $ in terms of $ \beta_0, \beta_1, \beta_2, \beta_3, \beta_4 $. We have now established that $ f(x) $ is a piecewise polynomial.

<span style="font-family: 'NewComputerModernMath';"> Since $ x > \xi $, this time the fifth term ($ \beta_4 (x - \xi)_+^3$) will be added to equation. So:

$$
\beta_4 (x - \xi)_+^3 = \beta_4x^3 -3\beta_4x^2\xi + 3\beta_4x\xi^2 -\beta_4\xi^3
$$


<span style="font-family: 'NewComputerModernMath';"> Therefore, 

$$
a_2 = \left (\beta_0  -\beta_4\xi^3  \right) ,\space b_2 =  \left (\beta_1 + 3\beta_4\xi^2  \right),\space c_2 = \left (\beta_2 -3\beta_4\xi   \right),\space d_2 = \left (\beta_3  +\beta_4 \right)
$$

<span style="font-family: 'NewComputerModernMath';">(c) Show that $ f_1(\xi) = f_2(\xi) $. That is, $ f(x) $ is continuous at $ \xi $.

$$
f_1(\xi) = \beta_0 + \beta_1 \xi + \beta_2 \xi^2 + \beta_3 \xi^3
$$

$$
f_2(\xi) = \beta_0 + \beta_1 \xi + \beta_2 \xi^2 + \beta_3 \xi^3 + \cancel{\beta_4 (\xi - \xi)^3}
$$

<span style="font-family: 'NewComputerModernMath';">Therefore,

$$
f_1(\xi) = f_2(\xi)
$$

<span style="font-family: 'NewComputerModernMath';">(d) Show that $ f_1'(\xi) = f_2'(\xi) $. That is, $ f'(x) $ is continuous at $ \xi $.

$$
f_1'(\xi) = \beta_1 + 2 \beta_2 x + 3 \beta_3 x^2  \rightarrow f_1'(\xi) = \beta_1 + 2 \beta_2 \xi + 3 \beta_3 \xi^2
$$

$$
f_2'(\xi) = \beta_1 + 2 \beta_2 x + 3 \beta_3 x^2 + 3 \beta_4 (x - \xi)_+^2 \rightarrow f_2'(\xi) =  \beta_1 + 2 \beta_2 \xi + 3 \beta_3 \xi^2 + \cancel{3 \beta_4 (\xi - \xi)_+^2}
$$

<span style="font-family: 'NewComputerModernMath';">Therefore,

$$
f_1'(\xi) = f_2'(\xi)
$$

<span style="font-family: 'NewComputerModernMath';">(e) Show that $ f_1''(\xi) = f_2''(\xi) $. That is, $ f''(x) $ is continuous at $ \xi $.

<span style="font-family: 'NewComputerModernMath';">Therefore, $ f(x) $ is indeed a cubic spline.

$$
f_1''(x) = 2 \beta_2 + 6 \beta_3 x  \rightarrow f_1''(\xi) = 2 \beta_2 + 6 \beta_3 \xi
$$

$$
f_2''(\xi) = 2 \beta_2 + 6 \beta_3 x + 6 \beta_4 (x - \xi)_+ \rightarrow f_2''(\xi) = 2 \beta_2 + 6 \beta_3 \xi + \cancel{6 \beta_4 (\xi - \xi)_+}
$$

<span style="font-family: 'NewComputerModernMath';">Therefore,

$$
f_1''(\xi) = f_2''(\xi)
$$

<span style="font-family: 'NewComputerModernMath';">*Hint:* Parts (d) and (e) of this problem require knowledge of single-variable calculus. As a reminder, given a cubic polynomial:

$$
f_1(x) = a_1 + b_1 x + c_1 x^2 + d_1 x^3,
$$

<span style="font-family: 'NewComputerModernMath';">the first derivative takes the form:

$$
f_1'(x) = b_1 + 2c_1 x + 3d_1 x^2
$$

<span style="font-family: 'NewComputerModernMath';">and second derivative takes the form:

$$
f_1''(x) = 2c_1 + 6d_1 x.

#### <span style="font-family: 'NewComputerModernMath';"> 3. Suppose we fit a curve with basis functions $ b_1(X) = X $, $ b_2(X) = (X - 1)^2 I(X \geq 1) $. (Note that $ I(X \geq 1) $ equals 1 for $ X \geq 1 $ and 0 otherwise.) We fit the linear regression model

$$
Y = \beta_0 + \beta_1 b_1(X) + \beta_2 b_2(X) + \epsilon,
$$

#### <span style="font-family: 'NewComputerModernMath';">and obtain coefficient estimates $ \hat{\beta}_0 = 1 $, $ \hat{\beta}_1 = 1 $, $ \hat{\beta}_2 = -2 $. Sketch the estimated curve between $ X = -2 $ and $ X = 2 $. Note the intercepts, slopes, and other relevant information.
</span>

```python
import numpy as np
import matplotlib.pyplot as plt

def b1(X):
    return X

def b2(X):
    return np.where(X >= 1, (X - 1)**2, 0)

def model(X, beta0, beta1, beta2):
    return beta0 + beta1 * b1(X) + beta2 * b2(X)

# Coefficients
beta0 = 1
beta1 = 1
beta2 = -2

X_values = np.linspace(-2, 2, 400)

Y_values = model(X_values, beta0, beta1, beta2)

plt.figure(figsize=(8, 6))
plt.plot(X_values, Y_values, label=r'', color='blue')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Estimated Curve')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)

plt.scatter(1, model(1, beta0, beta1, beta2), color='red', zorder=5, label='At X=1')

plt.show()
```

$$
Y = \beta_0 + \beta_1 b_1(X) + \beta_2 b_2(X) + \epsilon,
$$

#### <span style="font-family: 'NewComputerModernMath';"> and obtain coefficient estimates $ \hat{\beta}_0 = 1 $, $ \hat{\beta}_1 = 1 $, $ \hat{\beta}_2 = 3 $. Sketch the estimated curve between $ X = -2 $ and $ X = 6 $. Note the intercepts, slopes, and other relevant information.
</span>

```python
def b1(X):
    return np.where((X >= 0) & (X <= 2), 1, 0) - np.where((X >= 1) & (X <= 2), X - 1, 0)

def b2(X):
    return np.where((X >= 3) & (X <= 4), X - 3, 0) + np.where((X > 4) & (X <= 5), 1, 0)

def model(X, beta0, beta1, beta2):
    return beta0 + beta1 * b1(X) + beta2 * b2(X)

beta0 = 1
beta1 = 1
beta2 = 3

X_values = np.linspace(-2, 6, 400)

Y_values = model(X_values, beta0, beta1, beta2)

plt.figure(figsize=(8, 6))
plt.plot(X_values, Y_values, label='Estimated Curve', color='blue')
plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.title('Estimated Curve: Y = β₀ + β₁b₁(X) + β₂b₂(X)')
plt.xlabel('X')
plt.ylabel('Y')
plt.grid(True)
plt.legend()

plt.show()
```

#### <span style="font-family: 'NewComputerModernMath';"> (a) Perform polynomial regression to predict <span style="color: brown;">wage</span> using <span style="color: brown;">age</span>. Use cross-validation to select the optimal degree $ d $$ for the polynomial. What degree was chosen, and how does this compare to the results of hypothesis testing using ANOVA? Make a plot of the resulting polynomial fit to the data.

```python
X = Wage["age"].values.reshape(-1,1)
y = Wage["wage"].values.reshape(-1,1)
```

```python
degrees = np.arange(1, 15)  # Testing polynomial degrees from 1 to 10
# Function to manually create polynomial features using np.poly1d
def create_polynomial_features(X, degree):
    poly = PolynomialFeatures(degree)
    return poly.fit_transform(X)

# Cross-validation to select the optimal degree using OLS
cv_scores = []
kf = KFold(n_splits=10, shuffle=True, random_state=42)  # 10-fold cross-validation

for degree in degrees:
    fold_errors = []
    
    for train_idx, test_idx in kf.split(X):
        X_train, X_test = X[train_idx], X[test_idx]
        y_train, y_test = y[train_idx], y[test_idx]
        
        # Create polynomial features
        X_train_poly = create_polynomial_features(X_train, degree)
        X_test_poly = create_polynomial_features(X_test, degree)
        
        # Add constant term for OLS
        X_train_poly = sm.add_constant(X_train_poly)
        X_test_poly = sm.add_constant(X_test_poly)
        
        # Fit OLS model
        model = sm.OLS(y_train, X_train_poly).fit()
        
        # Make predictions and calculate mean squared error
        y_pred = model.predict(X_test_poly)
        mse = mean_squared_error(y_test, y_pred)
        fold_errors.append(mse)
    
    # Compute average MSE across all folds
    cv_scores.append(np.mean(fold_errors))

# Choose the optimal degree
optimal_degree = degrees[np.argmin(cv_scores)]

# Fit the optimal degree polynomial with OLS
X_poly = create_polynomial_features(X, optimal_degree)
X_poly = sm.add_constant(X_poly)  # Add constant term
model = sm.OLS(y, X_poly).fit()

# Make predictions
y_pred = model.predict(X_poly)
y_preds = model.get_prediction(X_poly)
bands = y_preds.conf_int(alpha=0.05)

# Plot the cross-validation scores for each polynomial degree
plt.figure(figsize=(10, 6))
plt.plot(degrees, cv_scores, color='blue', marker='o', label="Cross-validation error")
plt.axvline(optimal_degree, color='red', linestyle='--', label=f"Optimal Degree {optimal_degree}")
plt.title("Cross-validation Error for Polynomial Degrees")
plt.xlabel('Polynomial Degree')
plt.ylabel('Mean Squared Error (CV)')
plt.legend()
plt.grid(True)
plt.show()

sorted_idx = np.argsort(Wage["age"].values)
sorted_age = Wage["age"].values[sorted_idx]

plt.figure(figsize=(10, 6))
plt.scatter(X, y, label='Data')

# Loop over the predicted mean and bands, sorting each by age before plotting
for val, ls in zip([y_preds.predicted_mean, bands[:, 0], bands[:, 1]], ['b', 'r--', 'r--']):
    sorted_val = val[sorted_idx]
    plt.plot(sorted_age, sorted_val, ls, linewidth=3)

plt.xlabel("Age")
plt.ylabel("Wage")
plt.legend()
plt.title("Wage vs Age with Fitted Curve and Bands")
plt.show()

# Output the optimal degree and the corresponding cross-validation error
print("Optimal degree:", optimal_degree)
print("Cross-validation error:", np.min(cv_scores))
```

```python
anova_results = []
models=[]
for i in range(1, optimal_degree+1):  # Loop over polynomial degrees from 1 to 10
    # Generate polynomial features
    poly = PolynomialFeatures(degree=i)
    X_pol = poly.fit_transform(X)  # Polynomial features
    
    # Create a DataFrame with the features (column names as x1, x2, etc.)
    poly_df = pd.DataFrame(X_pol, columns=[f"x{i}" for i in range(X_pol.shape[1])])
    poly_df["wage"] = y  # Add the dependent variable
    
    # Fit the model using smf.ols
    formula = "wage ~ " + " + ".join(poly_df.columns[:-1])  # Create the formula
    model = smf.ols(formula, data=poly_df).fit()  # Fit the model using the formula
    
    models.append(model)  # Store the model for ANOVA

sm.stats.anova_lm(*models, typ=1)
```

<span style="font-family: 'NewComputerModernMath';">Lower F-values indicate less significance of the coefficients. Polynomial degrees greater than 4 do not significantly improve the model, which aligns with the cross-validation results.

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
import statsmodels.api as sm
import numpy as np
import matplotlib.pyplot as plt

target = Boston["nox"]

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(Boston, target, test_size=0.2, random_state=42)

# Extract the 'dis' and 'nox' columns (assuming 'dis' and 'nox' are features in the dataset)
X_train_dis = X_train[['dis']]
X_train_nox = X_train[['nox']]
X_test_dis = X_test[['dis']]
X_test_nox = X_test[['nox']]

# Create polynomial models for different degrees
models = [PolynomialFeatures(degree=d) for d in range(1, 11)]
Xs = [model.fit_transform(X_train_dis) for model in models]

# Calculate RSS for each polynomial model
rss_values = []

for X_ in Xs:
    model = sm.OLS(y_train, sm.add_constant(X_)).fit()  # Fit the OLS model (with intercept)
    predictions = model.predict(sm.add_constant(X_))  # Predict values using the fitted model
    residuals = y_train - predictions  # Residuals (actual - predicted)
    rss = np.sum(residuals**2)  # Residual sum of squares (RSS)
    rss_values.append(rss)

# Plot the RSS values
plt.figure(figsize=(8, 6))
plt.plot(range(1, len(rss_values) + 1), rss_values, marker='o', linestyle='-', color='b')
plt.title('Training Residual Sum of Squares (RSS) for Each Polynomial Model')
plt.xlabel('Polynomial Degree (Model Complexity)')
plt.ylabel('Residual Sum of Squares (RSS)')
plt.grid(True)
plt.show()
```

```python
kf = KFold(n_splits=5, shuffle=True, random_state=42)

# List to store average RSS for each polynomial degree
rss_values = []

# Loop over polynomial degrees from 1 to 11
for degree in range(1, 12):
    model = PolynomialFeatures(degree=degree)
    X_poly = model.fit_transform(X_train_dis)  # Transform the data using current degree
    
    fold_rss = []  # List to store RSS for each fold
    
    # Perform 10-fold cross-validation
    for train_index, val_index in kf.split(X_poly):
        X_train_fold, X_val_fold = X_poly[train_index], X_poly[val_index]
        y_train_fold, y_val_fold = y_train.iloc[train_index], y_train.iloc[val_index]
        
        # Fit the OLS model
        ols_model = sm.OLS(y_train_fold, sm.add_constant(X_train_fold)).fit()
        
        # Predict and calculate residuals
        predictions = ols_model.predict(sm.add_constant(X_val_fold))
        residuals = y_val_fold - predictions
        
        # Calculate the residual sum of squares (RSS) for this fold
        rss = np.sum(residuals**2)
        fold_rss.append(rss)
    
    # Average the RSS across all folds and append to the list
    avg_rss = np.mean(fold_rss)
    rss_values.append(avg_rss)

# Plot the average RSS values across different polynomial degrees
plt.figure(figsize=(8, 6))
plt.plot(range(1, 12), rss_values, marker='o', linestyle='-', color='b')
plt.title('Average Residual Sum of Squares (RSS) for Each Polynomial Degree (10-Fold CV)')
plt.xlabel('Polynomial Degree (Model Complexity)')
plt.ylabel('Average Residual Sum of Squares (RSS)')
plt.grid(True)
plt.show()

# Find the degree with the lowest RSS
best_degree = np.argmin(rss_values) + 1  # Adding 1 because degree starts from 1
print(f"The best polynomial degree based on 10-fold CV is: {best_degree}")
```

```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from itertools import combinations

# Define nCp function
def nCp(sigma2, estimator, X, Y):
    n, p = X.shape
    Yhat = estimator.predict(X)
    RSS = np.sum((Y - Yhat) ** 2)
    return -(RSS + 2 * p * sigma2) / n

# Forward Selection using nCp
def forward_selection_nCp(design, df,y):
    feature_names = list(design.transform(df).columns)
    selected_features = []
    Y = np.array(y)
    sigma2 = sm.OLS(Y, design.transform(df).values).fit().scale  # Compute variance
    
    best_nCp = float('-inf')

    while len(selected_features) < len(feature_names):
        best_feature = None
        for feature in feature_names:
            if feature not in selected_features:
                temp_features = selected_features + [feature]
                selected_indices = [feature_names.index(f) for f in temp_features]
                X_temp = design.transform(df).values[:, selected_indices]  # Ensure NumPy array indexing
                
                model_temp = sm.OLS(Y, X_temp).fit()
                nCp_temp = nCp(sigma2, model_temp, X_temp, Y)
                
                if nCp_temp > best_nCp:
                    best_nCp = nCp_temp
                    best_feature = feature

        if best_feature:
            selected_features.append(best_feature)
        else:
            break

    return selected_features

# Example Usage
design = MS(X_train).fit(X_train)  # Define design matrix
selected_features_forward = forward_selection_nCp(design, X_train, y_train)
print("Selected Features (Forward Selection with nCp):", selected_features_forward)
```

#### <span style="font-family: 'NewComputerModernMath';"> 12. This problem is a continuation of the previous exercise. In a toy  example with $ p = 100 $, show that one can approximate the multiple linear regression coefficient estimates by repeatedly performing simple linear regression in a backfitting procedure. How many backfitting iterations are required in order to obtain a “good” approximation to the multiple regression coefficient estimates? Create a plot to justify your answer.
</span>


```python
import numpy as np
import matplotlib.pyplot as plt
import statsmodels.api as sm

# Set the number of predictors and data points
n = 200  # number of samples
p = 100  # number of predictors

# Generate random data
np.random.seed(42)
X = np.random.rand(n, p)  # n x p matrix of predictors
true_beta = np.random.randint(-1000, 1000, p)
Y = X.dot(true_beta) + np.random.randn(n) * 0.1  # Y = X * beta + noise

# Function to perform backfitting procedure
def backfitting(X, Y, max_iter=100):
    n, p = X.shape
    beta_hat = np.zeros(p,)  # Initial guess of coefficients
    beta_history = np.zeros((max_iter, p))  # To track the coefficients over iterations
    error_history = np.zeros((max_iter, p))  # To track the error for each parameter over iterations
    total_error_history = np.zeros(max_iter)  # To track the total error over iterations

    for i in range(max_iter):
        residuals = Y.copy()
        # Update residuals by removing the effect of all other predictors
        for j in range(p):
            # Fit a model with all predictors except j
            X_others = np.delete(X, j, axis=1)
            residuals_tmp = residuals - X_others.dot(beta_hat[np.arange(p) != j])  # Remove effect of other predictors
            
            # Fit a model for the j-th predictor with the current residuals
            model = sm.OLS(residuals_tmp, X[:, j]).fit()
            beta_hat[j] = model.params[0]  # Update beta for predictor j
        
        beta_history[i, :] = beta_hat  # Store the coefficients

        # Calculate the error (absolute difference from the previous iteration)
        if i > 0:
            error_history[i, :] = np.abs(beta_history[i, :] - beta_history[i-1, :])

        # Calculate total error (sum of absolute errors across all parameters)
        total_error_history[i] = np.sum(error_history[i, :])

        # Check for convergence: if the coefficients don't change much, stop
        if i > 0 and np.linalg.norm(beta_hat - beta_history[i-1, :]) < 1e-6:
            break

    return beta_hat, beta_history, error_history, total_error_history, i + 1  # Return final coefficients, history, error history, and total error

# Perform backfitting
beta_hat, beta_history, error_history, total_error_history, num_iterations = backfitting(X, Y)

# Plot the error history for each parameter and the total error
colors = plt.cm.viridis(np.linspace(0, 1, error_history.shape[1]))

plt.figure(figsize=(10, 6))

# Loop through each parameter (each column) in error_history and plot with a unique color
for i in range(error_history.shape[1]):
    plt.plot(error_history[:, i], color=colors[i], alpha=0.6)

# Plot the total error line in red and thicker
plt.plot(total_error_history/p, color='red', linewidth=3, label='Average Error')

plt.xlabel('Iteration')
plt.ylabel('Error (Change in Parameter Value)')
plt.title('Error Rate for Each Parameter per Iteration')
plt.legend(loc='upper right')
plt.show()

# Optionally, print a few of the error values at the last few iterations
print(error_history[-1, :10])  # Show error rates for first 10 parameters at the last iteration
print(f"Total error at final iteration: {total_error_history[-1]}")
```
