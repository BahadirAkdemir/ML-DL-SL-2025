![Data Set](e2_4-18.png)



##### <span style="font-family: 'NewComputerModernMath';">To get the equation in 4.18, we need to manipulate 4.17 with several steps</span>

##### **i.** <span style="font-family: 'NewComputerModernMath';">In 4.17, the denominator is same for all classes. So, it will not affect the result while choosing the maximum value. We can remove it. We get </span>

$$
p_k(x) = \frac{\pi_k \frac{1}{\sqrt{2 \pi} \sigma} \exp \left( -\frac{1}{2 \sigma^2} (x - \mu_k)^2 \right)}{\sum_{l=1}^{K} \pi_l \frac{1}{\sqrt{2 \pi} \sigma} \exp \left( -\frac{1}{2 \sigma^2} (x - \mu_l)^2 \right)} \overset{\text{removing denominator}}{\rightarrow} \pi_k \frac{1}{\sqrt{2 \pi} \sigma} \exp \left( -\frac{1}{2 \sigma^2} (x - \mu_k)^2 \right) 
$$

##### **ii.** <span style="font-family: 'NewComputerModernMath';">We can take the logarithm of the remain part due to the fact that logarithm has no effect in choosing maximum value</span> $(\log n  > \log k \space \text{if} \space n > k)$. <span style="font-family: 'NewComputerModernMath';">After that, we can remove the parameters that has no unknown variables because these also have no affect on finding maximum. Now we get</span>

$$
 \pi_k \frac{1}{\sqrt{2 \pi} \sigma} \exp  \left( -\frac{1}{2 \sigma^2} (x - \mu_k)^2 \right) \overset{\text{taking logarithm}}{\rightarrow} \log{\pi_k} -  \log{\sqrt{2 \pi} \sigma} -\frac{1}{2 \sigma^2} (x - \mu_k)^2 \overset{\text{removing unnecassary parameters}}{\rightarrow} \log{\pi_k} -\frac{1}{2 \sigma^2} (x - \mu_k)^2
$$

##### **iii.** <span style="font-family: 'NewComputerModernMath';">After expanding square terms, dropping the values that not depend on class k (e.g. x variable will be same for all classes so we can also drop them), we get</span>

$$
\log{\pi_k} -\frac{1}{2 \sigma^2} (x - \mu_k)^2 \overset{\text{expanding terms}}{\rightarrow} \log{\pi_k} -\frac{1}{2 \sigma^2} (x^2 - 2 x \mu_k + \mu_k^2) = \log{\pi_k} -\frac{x^2}{2 \sigma^2} +\frac{\cancel{2} x \mu_k}{\cancel{2} \sigma^2}- \frac{\mu_k^2}{2 \sigma^2} \overset{\text{dropping unnecassry terms}}{\rightarrow} \log{\pi_k} + \frac{\mu_k}{\sigma^2}- \frac{\mu_k^2}{2 \sigma^2} = \delta_{k} (x)
$$

##### <span style="font-family: 'NewComputerModernMath';">Finally, we get a new formula that helps to find the same solution with less computation.</span>

#### <span style="font-family: 'NewComputerModernMath', serif;"> 11. Work out the detailed forms of $( a_k )$, $( b_{kj} )$, and $( b_{kjl} )$ in  <a href="#4.33" style="color: blue;">(4.33)</a>. Your answer should involve  $( \pi_k )$, $( \pi_K )$, $( \mu_k )$, $( \mu_K )$, $( \boldsymbol{\Sigma}_k )$, and $( \boldsymbol{\Sigma}_K )$.
</span>


##### <span style="font-family: 'NewComputerModernMath', serif;">Question is similar to the previous one, but this time we will assume that $p \ge 1$</span>

$$
\begin{align}
\log \left( \frac{\Pr(Y = k | X = x)}{\Pr(Y = K | X = x)} \right) 
&= \log \left( \frac{\pi_k f_k(x)}{\pi_K f_K(x)} \right) \\
&= \log \left( \frac{\pi_k \times e^{\left( -\frac{1}{2} (x - \mu_k)^T \Sigma_k^{-1} (x - \mu_k) \right)}}{\pi_K \times e^{\left( -\frac{1}{2} (x - \mu_K)^T \Sigma_K^{-1} (x - \mu_K) \right)}} \right) \\
&= \log \left( \frac{\pi_k}{\pi_K} \right) - \left(\frac{1}{2} (x - \mu_k)^T \Sigma_k^{-1} (x - \mu_k) \right) + \left(\frac{1}{2} (x - \mu_K)^T \Sigma_K^{-1} (x - \mu_K) \right) \\
&= \log \left( \frac{\pi_k}{\pi_K} \right)  - \frac{1}{2} \left(  (x - \mu_k)^T \Sigma_k^{-1} (x - \mu_k)  - (x - \mu_K)^T \Sigma_K^{-1} (x - \mu_K) \right)
\end{align}
$$

<span style="font-family: 'NewComputerModernMath', serif;"> The quadratic terms in the expression can be expanded as follows:</span>

<span style="font-family: 'NewComputerModernMath', serif;">1. For class</span> <span style="font-family: 'NewComputerModernMath';"><i>k</i></span>:

$$
(x - \mu_k)^T \Sigma_k^{-1} (x - \mu_k) = x^T \Sigma_k^{-1} x - 2\mu_k^T \Sigma_k^{-1} x + \mu_k^T \Sigma_k^{-1} \mu_k.
$$

<span style="font-family: 'NewComputerModernMath', serif;">2. For class</span> <span style="font-family: 'NewComputerModernMath';"><i>K</i></span>:

   $$
   (x - \mu_K)^T \Sigma_K^{-1} (x - \mu_K) = x^T \Sigma_K^{-1} x - 2\mu_K^T \Sigma_K^{-1} x + \mu_K^T \Sigma_K^{-1} \mu_K.
   $$

<span style="font-family: 'NewComputerModernMath', serif;"> Substituting these into the log-ratio, we get:</span>

$$
\log \left( \frac{\Pr(Y = k | X = x)}{\Pr(Y = K | X = x)} \right) = \log \left( \frac{\pi_k}{\pi_K} \right) - \frac{1}{2} \left( x^T (\Sigma_k^{-1} - \Sigma_K^{-1}) x - 2 (\mu_k^T \Sigma_k^{-1} - \mu_K^T \Sigma_K^{-1}) x + (\mu_k^T \Sigma_k^{-1} \mu_k - \mu_K^T \Sigma_K^{-1} \mu_K) \right)
$$

<span style="font-family: 'NewComputerModernMath', serif;">The expression can be rewritten in the form:</span>

$$
\log \left( \frac{\Pr(Y = k \mid X = x)}{\Pr(Y = K \mid X = x)} \right) = a_k + \sum_{j=1}^{p} b_{kj} x_j + \sum_{j=1}^{p} \sum_{l=1}^{p} b_{kjl} x_j x_l,
$$

<span style="font-family: 'NewComputerModernMath', serif;"> where:</span>

- <span style="font-family: 'NewComputerModernMath';">$( a_k )$</span> is the constant term,
- <span style="font-family: 'NewComputerModernMath';">$( b_{kj} )$</span> is the coefficient of the linear term <span style="font-family: 'NewComputerModernMath';">$( x_j )$</span>,
- <span style="font-family: 'NewComputerModernMath';">$( b_{kjl} )$</span> is the coefficient of the quadratic term <span style="font-family: 'NewComputerModernMath';">$( x_j x_l )$</span>.

<span style="font-family: 'NewComputerModernMath', serif;"> By matching terms, we obtain:</span>

##### <span style="font-family: 'NewComputerModernMath';">1. **Constant Term** ($ a_k $):</span>

$$
a_k = \log \left( \frac{\pi_k}{\pi_K} \right) - \frac{1}{2} \left( \mu_k^T \Sigma_k^{-1} \mu_k - \mu_K^T \Sigma_K^{-1} \mu_K \right).
$$

#### <span style="font-family: 'NewComputerModernMath';">2. **Linear Term** ($ b_{kj} $):</span>

<span style="font-family: 'NewComputerModernMath';">The linear term arises from the expression  $ -2 (\mu_k^T \Sigma_k^{-1} - \mu_K^T \Sigma_K^{-1}) x $. The coefficient $ b_{kj} $ is:</span>

$$
b_{kj} = (\mu_K^T \Sigma_K^{-1} - \mu_k^T \Sigma_k^{-1})_j,
$$

<span style="font-family: 'NewComputerModernMath';">where $ (\cdot)_j $denotes the jth component of the vector </span>

#### <span style="font-family: 'NewComputerModernMath';">3. **Quadratic Term** ($( b_{kjl} )$):</span>

<span style="font-family: 'NewComputerModernMath';">The quadratic term arises from the expression  $( x^T (\Sigma_k^{-1} - \Sigma_K^{-1}) x )$. The coefficient $b_{kjl}$ is:</span>

$$
b_{kjl} = -\frac{1}{2} \left( (\Sigma_k^{-1} - \Sigma_K^{-1})_{jl} \right),
$$

<span style="font-family: 'NewComputerModernMath';">where $( (\cdot)_{jl} )$ denotes the $( (j, l) )$th entry of the matrix.</span>

<span style="font-family: 'NewComputerModernMath';"> 1. Logistic Regression with All Combinations & Interactions

`````python
import itertools
from itertools import combinations
from sklearn.preprocessing import PolynomialFeatures

best_logit_acc = 0
best_logit_model = None
best_logit_vars = None


base_predictors = ['Lag1', 'Lag2', 'Lag3', 'Lag4', 'Lag5', 'Volume']


# Iterate over all feature combinations
for r in range(1, len(base_predictors) + 1):
    for subset in combinations(base_predictors, r):
        # Create polynomial features (interactions)
        poly = PolynomialFeatures(degree=3, interaction_only=True, include_bias=False)
        X_train_poly = poly.fit_transform(train_Weekly[list(subset)])
        X_test_poly = poly.transform(test_Weekly[list(subset)])
        
        # Fit logistic regression
        X_train_poly = sm.add_constant(X_train_poly)
        X_test_poly = sm.add_constant(X_test_poly)
        y_train = train_Weekly["Direction"]

        logit_model = sm.Logit(y_train, X_train_poly)
        logit_result = logit_model.fit(disp=0)  # Suppress output

        # Predictions
        logit_pred = (logit_result.predict(X_test_poly) > 0.5).astype(int)
        acc = np.mean(logit_pred == test_Weekly["Direction"])

        # Store best performing model
        if acc > best_logit_acc:
            best_logit_acc = acc
            best_logit_model = logit_result
            best_logit_vars = subset

print(f"Best Logistic Regression Model: {best_logit_vars} with accuracy {best_logit_acc:.4f}")
`````
