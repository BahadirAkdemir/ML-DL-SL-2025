````python
import numpy as np
import matplotlib.pyplot as plt

# Define the hyperplane equation X2 = 3X1 + 1
X1 = np.linspace(-2, 2, 100)
X2 = 3 * X1 + 1

# Create the plot
plt.figure(figsize=(7, 5))
plt.plot(X1, X2, 'b-', label=r'$1 + 3X_1 - X_2 = 0$')

# Shade the regions
plt.fill_between(X1, X2, max(X2) + 1, color='lightblue', alpha=0.5, label=r'$1 + 3X_1 - X_2 > 0$ (above)')
plt.fill_between(X1, X2, min(X2) - 1, color='lightcoral', alpha=0.5, label=r'$1 + 3X_1 - X_2 < 0$ (below)')

# Plot intercepts
plt.scatter([-1/3, 0], [0, 1], color='red', zorder=3, label='Intercepts')

# Labels and formatting
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.xlim(-2, 2)
plt.ylim(-3, 3)
plt.legend()
plt.title("Sketch of the Hyperplane")

# Show the plot
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````

````python
# Define the circle parameters
center_x, center_y = -1, 2
radius = 2

# Generate points for the circle
theta = np.linspace(0, 2*np.pi, 300)
X1_circle = center_x + radius * np.cos(theta)
X2_circle = center_y + radius * np.sin(theta)

# Create the plot
plt.figure(figsize=(6, 6))
plt.plot(X1_circle, X2_circle, 'b-', label=r'$(1 + X_1)^2 + (2 - X_2)^2 = 4$')

# Mark the center
plt.scatter(center_x, center_y, color='red', zorder=3, label='Center (-1,2)')

# Labels and formatting
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
plt.xlabel(r'$X_1$')
plt.ylabel(r'$X_2$')
plt.xlim(-4, 2)
plt.ylim(-1, 5)
plt.legend()
plt.title("Sketch of the Circle")

# Show the plot
plt.grid(True, linestyle='--', linewidth=0.5)
plt.show()
````

## <span style="color: rgb(0, 105, 175);font-family: 'NewComputerModernMath';">*Applied*</span>

### <span style="color: rgb(0, 105, 175);font-family: 'NewComputerModernMath';">Q4</span>

#### <span style="font-family: 'NewComputerModernMath';">4. Generate a simulated two-class data set with 100 observations and two features in which there is a visible but non-linear separation between the two classes. Show that in this setting, a support vector machine with a polynomial kernel (with degree greater than 1) or a radial kernel will outperform a support vector classifier on the training data. Which technique performs best on the test data? Make plots and report training and test error rates in order to back up your assertions.
</span>

````python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Generate non-linearly separable data
np.random.seed(42)
n_samples = 100

X1 = np.random.randn(n_samples // 2, 2) - [1.5, 1.5]
X2 = np.random.randn(n_samples // 2, 2) + [1.5, 1.5]
X = np.vstack((X1, X2))
y = np.hstack((np.zeros(n_samples // 2), np.ones(n_samples // 2)))






X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)







svc_linear = SVC(kernel="linear", C=1).fit(X_train, y_train)
svc_poly = SVC(kernel="poly", degree=3, C=1).fit(X_train, y_train)
svc_rbf = SVC(kernel="rbf", C=1).fit(X_train, y_train)



y_train_pred_linear = svc_linear.predict(X_train)
y_test_pred_linear = svc_linear.predict(X_test)

y_train_pred_poly = svc_poly.predict(X_train)
y_test_pred_poly = svc_poly.predict(X_test)

y_train_pred_rbf = svc_rbf.predict(X_train)
y_test_pred_rbf = svc_rbf.predict(X_test)



train_acc_linear = accuracy_score(y_train, y_train_pred_linear)
test_acc_linear = accuracy_score(y_test, y_test_pred_linear)

train_acc_poly = accuracy_score(y_train, y_train_pred_poly)
test_acc_poly = accuracy_score(y_test, y_test_pred_poly)

train_acc_rbf = accuracy_score(y_train, y_train_pred_rbf)
test_acc_rbf = accuracy_score(y_test, y_test_pred_rbf)

# Plot decision boundaries
def plot_decision_boundary(model, X, y, title):
    h = 0.02
    x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
    y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
    xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
    
    Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)
    
    plt.contourf(xx, yy, Z, alpha=0.3)
    plt.scatter(X[:, 0], X[:, 1], c=y, edgecolors='k')
    plt.title(title)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plot_decision_boundary(svc_linear, X_train, y_train, "SVM with Linear Kernel (Train)")

plt.subplot(1, 3, 2)
plot_decision_boundary(svc_poly, X_train, y_train, "SVM with Polynomial Kernel (Train)")

plt.subplot(1, 3, 3)
plot_decision_boundary(svc_rbf, X_train, y_train, "SVM with RBF Kernel (Train)")

plt.show()

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plot_decision_boundary(svc_linear, X_test, y_test, "SVM with Linear Kernel (Test)")

plt.subplot(1, 3, 2)
plot_decision_boundary(svc_poly, X_test, y_test, "SVM with Polynomial Kernel (Test)")

plt.subplot(1, 3, 3)
plot_decision_boundary(svc_rbf, X_test, y_test, "SVM with RBF Kernel (Test)")

plt.show()

# Report results
results = {
    "Model": ["Linear SVM", "Polynomial SVM", "RBF SVM"],
    "Train Accuracy": [train_acc_linear, train_acc_poly, train_acc_rbf],
    "Test Accuracy": [test_acc_linear, test_acc_poly, test_acc_rbf]
}

results_df = pd.DataFrame(results)
results_df
````
